```python
#获取 2011-2020年的数据

hp2011_df = pd.read_csv("2011.csv")

hp2012_df = pd.read_csv("2012.csv")

hp2013_df = pd.read_csv("2013.csv")

hp2014_df = pd.read_csv("2014.csv")

hp2015_df = pd.read_csv("2015.csv")

hp2016_df = pd.read_csv("2016.csv")

hp2017_df = pd.read_csv("2017.csv")

hp2018_df = pd.read_csv("2018.csv")

hp2019_df = pd.read_csv("2019.csv")

hp2020_df = pd.read_csv("2020.csv")

#进行堆叠拼接，注意：多个表格的拼接传入元组或列表 all_hour_pri命名DF

all_hou_pri= pd.concat((hp2011_df,hp2012_df,hp2013_df,hp2014_df,hp2015_df,hp2016_df,
           hp2017_df,hp2018_df,hp2019_df,hp2020_df))
```

```python
all_hou_pri
```

```python
all_hou_pri.columns
```

```python
# can drop some useless columns

to_drop = ['Unnamed: 0', 'raw_data', 'percent', 'upordown']

all_hou_pri.drop(columns=to_drop, inplace=True)
```

```python
all_hou_pri.columns
```

```python
all_hou_pri.head()
```

```python
all_hou_pri['date'] = all_hou_pri['date'].apply(lambda s : s[:-3])
```

```python
all_hou_pri.head()
```

```python
all_hou_pri['price'] = all_hou_pri['price'].apply(lambda s : s[:-2])
```

```python
all_hou_pri.head()
all_hou_pri['year'] = all_hou_pri['date'].str.extract(r'^(\d{4})', expand=False) 
all_hou_pri['month'] = all_hou_pri['date'].str.extract(r'(\d{2}$)', expand=False)
all_hou_pri['year_month'] = all_hou_pri['year']+'-'+ all_hou_pri['month']
```
